# Solução de PVCs com o módulo FEniCS (módulo python para elementos finitos)

```python
"""
Resolução de Problemas de Valor de Contorno com Condições de Neumann usando FEniCS

Instalação do FEniCS:
---------------------
# Opção 1: Usando conda (recomendado)
conda create -n fenics
conda activate fenics
conda install -c conda-forge fenics

# Opção 2: Usando Docker
docker run -ti -v $(pwd):/home/fenics/shared quay.io/fenicsproject/stable

# Opção 3: Usando apt (Ubuntu/Debian)
sudo apt-get install fenics
"""

from fenics import *
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Configurar para não exibir logs desnecessários
set_log_level(LogLevel.WARNING)

def solve_1d_neumann():
    """
    Resolve -u''(x) = f(x) em [0, 1]
    com u'(0) = g_L e u'(1) = g_R
    """
    print("=== Problema 1D com Condições de Neumann ===\n")

    # Criar malha 1D
    mesh = IntervalMesh(100, 0, 1) # 100 pontos de igualmente espaçados indo de 0 a 1

    '''
      Definir espaço de elementos finitos (P1 = polinômios lineares)
      esse é o espaço onde estarão as "funções de base", que serão utilizadas para a aproximação do método
    '''

    V = FunctionSpace(mesh, 'P', 1)

    # Definir as condições de Neumann
    g_L = Constant(0.0)  # u'(0) = 0
    g_R = Constant(0.0)  # u'(1) = 0

    # Definir termo fonte
    f = Expression('sin(2*pi*x[0])', degree=2)

    # Definir a solução teste e trial
    u = TrialFunction(V)
    v = TestFunction(V)

    # Formulação variacional
    # ∫ ∇u·∇v dx = ∫ f·v dx + ∫_∂Ω g·v ds
    a = dot(grad(u), grad(v))*dx
    L = f*v*dx

    # Adicionar termos de Neumann nas fronteiras
    # Definir as fronteiras
    def boundary_left(x, on_boundary):
        return on_boundary and near(x[0], 0.0)

    def boundary_right(x, on_boundary):
        return on_boundary and near(x[0], 1.0)

    # Criar medidas para integração nas fronteiras
    boundary_markers = MeshFunction("size_t", mesh, mesh.topology().dim()-1, 0) # usado para criar índices para as fronteiras

    class LeftBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return boundary_left(x, on_boundary)

    class RightBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return boundary_right(x, on_boundary)

    left = LeftBoundary()
    right = RightBoundary()

    left.mark(boundary_markers, 1) # A fronteira esquerda recebe o índice 1 e a esquerda o índice 2
    right.mark(boundary_markers, 2)

    # Redefinir medida ds para usar marcadores
    ds = Measure('ds', domain=mesh, subdomain_data=boundary_markers)

    # Adicionar condições de Neumann ao funcional linear
    L = f*v*dx + g_L*v*ds(1) + g_R*v*ds(2)

    # Como temos apenas condições de Neumann, precisamos fixar um ponto
    # para garantir unicidade. Vamos impor u(0.5) = 0

    # Criar uma condição de Dirichlet em um único ponto
    def midpoint(x, on_boundary):
        return near(x[0], 0.5)

    bc = DirichletBC(V, Constant(0.0), midpoint, method='pointwise')

    # Resolver o sistema
    u = Function(V)
    solve(a == L, u, bc)

    # Plotar solução
    plt.figure(figsize=(12, 5))

    # Extrair valores para plotagem
    x_vals = mesh.coordinates()[:, 0]
    u_vals = u.compute_vertex_values()

    # Ordenar por x (necessário para plotagem correta)
    idx = np.argsort(x_vals)
    x_vals = x_vals[idx]
    u_vals = u_vals[idx]

    plt.subplot(1, 2, 1)
    plt.plot(x_vals, u_vals, 'b-', linewidth=2, label='Solução u(x)')
    plt.xlabel('x')
    plt.ylabel('u(x)')
    plt.title('Solução 1D com Condições de Neumann (FEniCS)')
    plt.grid(True, alpha=0.3)
    plt.legend()

    # Calcular e plotar derivada
    V_grad = FunctionSpace(mesh, 'P', 1)
    u_grad = project(grad(u)[0], V_grad)
    grad_vals = u_grad.compute_vertex_values()[idx]

    plt.subplot(1, 2, 2)
    plt.plot(x_vals, grad_vals, 'r-', linewidth=2, label="u'(x)")
    plt.axhline(y=float(g_L), color='g', linestyle='--', label=f"u'(0) = {float(g_L)}")
    plt.axhline(y=float(g_R), color='g', linestyle='--', label=f"u'(1) = {float(g_R)}")
    plt.xlabel('x')
    plt.ylabel("u'(x)")
    plt.title('Verificação das Condições de Neumann')
    plt.grid(True, alpha=0.3)
    plt.legend()

    plt.tight_layout()
    plt.show()

    return u

def solve_2d_neumann():
    """
    Resolve -∆u = f em Ω = [0,1]×[0,1]
    com ∂u/∂n = g em toda a fronteira
    """
    print("\n=== Problema 2D com Condições de Neumann ===\n")

    # Criar malha 2D
    mesh = RectangleMesh(Point(0, 0), Point(1, 1), 30, 30)

    # Espaço de elementos finitos
    V = FunctionSpace(mesh, 'P', 1)

    # Termo fonte
    f = Expression('2*sin(pi*x[0])*sin(pi*x[1])', degree=2)

    # Condições de Neumann em cada parte da fronteira
    g_left = Constant(0.0)   # ∂u/∂n = 0 em x=0
    g_right = Constant(0.0)  # ∂u/∂n = 0 em x=1
    g_bottom = Constant(0.0) # ∂u/∂n = 0 em y=0
    g_top = Constant(0.0)    # ∂u/∂n = 0 em y=1

    # Marcar diferentes partes da fronteira
    boundary_markers = MeshFunction("size_t", mesh, mesh.topology().dim()-1, 0)

    class LeftBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and near(x[0], 0.0)

    class RightBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and near(x[0], 1.0)

    class BottomBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and near(x[1], 0.0)

    class TopBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and near(x[1], 1.0)

    # Marcar fronteiras
    LeftBoundary().mark(boundary_markers, 1)
    RightBoundary().mark(boundary_markers, 2)
    BottomBoundary().mark(boundary_markers, 3)
    TopBoundary().mark(boundary_markers, 4)

    # Definir medida de integração na fronteira
    ds = Measure('ds', domain=mesh, subdomain_data=boundary_markers)

    # Formulação variacional
    u = TrialFunction(V)
    v = TestFunction(V)

    a = dot(grad(u), grad(v))*dx
    L = f*v*dx + g_left*v*ds(1) + g_right*v*ds(2) + g_bottom*v*ds(3) + g_top*v*ds(4)

    # Fixar um ponto para unicidade (centro do domínio)
    def center(x, on_boundary):
        return near(x[0], 0.5) and near(x[1], 0.5)

    bc = DirichletBC(V, Constant(0.0), center, method='pointwise')

    # Resolver
    u = Function(V)
    solve(a == L, u, bc)

    # Plotar solução
    fig = plt.figure(figsize=(14, 6))

    # Extrair dados para plotagem
    x = mesh.coordinates()[:, 0]
    y = mesh.coordinates()[:, 1]
    u_vals = u.compute_vertex_values()

    # Criar grade regular para plotagem
    from scipy.interpolate import griddata
    xi = np.linspace(0, 1, 100)
    yi = np.linspace(0, 1, 100)
    Xi, Yi = np.meshgrid(xi, yi)
    Zi = griddata((x, y), u_vals, (Xi, Yi), method='cubic')

    # Contour plot
    ax1 = fig.add_subplot(1, 2, 1)
    contour = ax1.contourf(Xi, Yi, Zi, levels=20, cmap='viridis')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_title('Solução 2D com Condições de Neumann (FEniCS)')
    plt.colorbar(contour, ax=ax1, label='u(x,y)')

    # 3D surface plot
    ax2 = fig.add_subplot(1, 2, 2, projection='3d')
    surf = ax2.plot_surface(Xi, Yi, Zi, cmap='viridis', alpha=0.9)
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('u(x,y)')
    ax2.set_title('Visualização 3D')
    plt.colorbar(surf, ax=ax2, shrink=0.5)

    plt.tight_layout()
    plt.show()

    return u

def solve_mixed_boundary():
    """
    Exemplo com condições mistas: Dirichlet em parte da fronteira,
    Neumann no resto
    """
    print("\n=== Problema com Condições Mistas (Dirichlet + Neumann) ===\n")

    # Criar malha
    mesh = UnitSquareMesh(40, 40)
    V = FunctionSpace(mesh, 'P', 1)

    # Definir partes da fronteira
    def dirichlet_boundary(x, on_boundary):
        # Dirichlet em x=0 e x=1
        return on_boundary and (near(x[0], 0.0) or near(x[0], 1.0))

    # Condição de Dirichlet
    u_D = Expression('x[1]*(1-x[1])', degree=2)
    bc = DirichletBC(V, u_D, dirichlet_boundary)

    # Termo fonte
    f = Constant(1.0)

    # Condições de Neumann em y=0 e y=1
    g_bottom = Expression('0.5*x[0]*(1-x[0])', degree=2)  # Fluxo não-zero
    g_top = Expression('-0.5*x[0]*(1-x[0])', degree=2)    # Fluxo oposto

    # Marcar fronteiras para Neumann
    boundary_markers = MeshFunction("size_t", mesh, mesh.topology().dim()-1, 0)

    class BottomBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and near(x[1], 0.0)

    class TopBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and near(x[1], 1.0)

    BottomBoundary().mark(boundary_markers, 1)
    TopBoundary().mark(boundary_markers, 2)

    ds = Measure('ds', domain=mesh, subdomain_data=boundary_markers)

    # Formulação variacional
    u = TrialFunction(V)
    v = TestFunction(V)

    a = dot(grad(u), grad(v))*dx
    L = f*v*dx + g_bottom*v*ds(1) + g_top*v*ds(2)

    # Resolver
    u = Function(V)
    solve(a == L, u, bc)

    # Plotar
    plt.figure(figsize=(10, 8))

    # Preparar dados
    x = mesh.coordinates()[:, 0]
    y = mesh.coordinates()[:, 1]
    u_vals = u.compute_vertex_values()

    from scipy.interpolate import griddata
    xi = np.linspace(0, 1, 100)
    yi = np.linspace(0, 1, 100)
    Xi, Yi = np.meshgrid(xi, yi)
    Zi = griddata((x, y), u_vals, (Xi, Yi), method='cubic')

    # Plotar solução
    contour = plt.contourf(Xi, Yi, Zi, levels=20, cmap='coolwarm')
    plt.colorbar(contour, label='u(x,y)')

    # Adicionar setas indicando fluxo (gradiente)
    skip = 5
    U = np.gradient(Zi, axis=1)[::skip, ::skip]
    V = np.gradient(Zi, axis=0)[::skip, ::skip]
    plt.quiver(Xi[::skip, ::skip], Yi[::skip, ::skip], U, V,
               alpha=0.5, color='black', scale=20)

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Problema Misto: Dirichlet (x=0,1) + Neumann (y=0,1)')

    # Adicionar anotações
    plt.text(0.5, -0.05, 'Neumann: ∂u/∂n = g_bottom', ha='center', transform=plt.gca().transAxes)
    plt.text(0.5, 1.02, 'Neumann: ∂u/∂n = g_top', ha='center', transform=plt.gca().transAxes)
    plt.text(-0.15, 0.5, 'Dirichlet', rotation=90, va='center', transform=plt.gca().transAxes)
    plt.text(1.05, 0.5, 'Dirichlet', rotation=90, va='center', transform=plt.gca().transAxes)

    plt.tight_layout()
    plt.show()

    return u

def solve_nonlinear_neumann():
    """
    Exemplo de problema não-linear com condições de Neumann
    -∇·(k(u)∇u) = f com k(u) = 1 + u²
    """
    print("\n=== Problema Não-Linear com Condições de Neumann ===\n")

    mesh = UnitSquareMesh(32, 32)
    V = FunctionSpace(mesh, 'P', 1)

    # Condição inicial
    u = Function(V)
    v = TestFunction(V)

    # Coeficiente não-linear
    def k(u):
        return 1 + u**2

    # Termo fonte
    f = Constant(1.0)

    # Condições de Neumann (zero em toda fronteira)
    g = Constant(0.0)

    # Formulação variacional não-linear
    F = k(u)*dot(grad(u), grad(v))*dx - f*v*dx - g*v*ds

    # Fixar um ponto para unicidade
    def center(x, on_boundary):
        return near(x[0], 0.5) and near(x[1], 0.5)

    bc = DirichletBC(V, Constant(0.0), center, method='pointwise')

    # Resolver problema não-linear
    solve(F == 0, u, bc)

    # Plotar
    plt.figure(figsize=(12, 5))

    # Preparar dados
    x = mesh.coordinates()[:, 0]
    y = mesh.coordinates()[:, 1]
    u_vals = u.compute_vertex_values()

    from scipy.interpolate import griddata
    xi = np.linspace(0, 1, 100)
    yi = np.linspace(0, 1, 100)
    Xi, Yi = np.meshgrid(xi, yi)
    Zi = griddata((x, y), u_vals, (Xi, Yi), method='cubic')

    # Contour plot
    plt.subplot(1, 2, 1)
    contour = plt.contourf(Xi, Yi, Zi, levels=20, cmap='plasma')
    plt.colorbar(contour, label='u(x,y)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Solução do Problema Não-Linear')

    # Plot do coeficiente k(u)
    plt.subplot(1, 2, 2)
    k_vals = 1 + u_vals**2
    Ki = griddata((x, y), k_vals, (Xi, Yi), method='cubic')
    contour2 = plt.contourf(Xi, Yi, Ki, levels=20, cmap='viridis')
    plt.colorbar(contour2, label='k(u) = 1 + u²')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Coeficiente de Difusão k(u)')

    plt.tight_layout()
    plt.show()

    return u

# Executar exemplos
if __name__ == "__main__":
    print("=" * 60)
    print("EXEMPLOS DE PROBLEMAS COM CONDIÇÕES DE NEUMANN USANDO FENICS")
    print("=" * 60)

    # Exemplo 1D
    u_1d = solve_1d_neumann()

    # Exemplo 2D
    u_2d = solve_2d_neumann()

    # Exemplo com condições mistas
    u_mixed = solve_mixed_boundary()

    # Exemplo não-linear
    u_nonlinear = solve_nonlinear_neumann()

    print("\n" + "=" * 60)
    print("Todos os exemplos foram executados com sucesso!")
    print("=" * 60)
```

## Workflow

1. Criar malha no domínio (já definindo o domínio)
2. Definir as condições de contorno
    1. As condições de contorno de Neumann entram dentro do termo linear do problema variacional.
        * Ex: ∫ ∇u·∇v dx = ∫ f·v dx + ∫_∂Ω g·v ds
            * O termo ∫_∂Ω g·v ds é o termo linear que contém as condições de Neumann imposta sobre as superfícies do domínio Ω.
3. Definir o espaço das funções de base
4. Definir as condições de contorno
5. Definir as funções de "trial" (a que irá representar a solução do PVC) e "test".
6. Formulação variacional do problema
7. Indexar fronteiras para definir BCs
8. Resolver PVC

1. Formulação variacional do problema

  * Formulação original
$$
  -\nabla\cdot\boldsymbol{\sigma}^{\varepsilon} = 0, \quad \mathbf{x} \in \Omega, \nonumber\\
  \boldsymbol{\sigma}^{\varepsilon} =
  \lambda^{\varepsilon} \, \mathrm{tr}(\boldsymbol{\epsilon}^{\varepsilon}) \, I + 2\mu^{\varepsilon} \boldsymbol{\epsilon}^{\varepsilon}, \quad
  \boldsymbol{\epsilon}^{\varepsilon} = \frac{1}{2}\left[\nabla \mathbf{u}^{\varepsilon} + (\nabla \mathbf{u}^{\varepsilon})^T\right], \\
  \boldsymbol{\sigma}^{\varepsilon} \cdot \mathbf{n} =
   -\mathbf{p}, \quad \mathbf{x} \in \Gamma_u, \quad
  \boldsymbol{\sigma}^{\varepsilon} \cdot \mathbf{n} =
   \mathbf{0}, \quad \mathbf{x} \in \Gamma_t.
$$

onde todas as funções são funções da variável macroscópica $(\mathbf{x})$.

  * Formulação variacional de problema de equilíbrio linear elástico

$$
  \int_{\Omega}  \sigma(\mathbf{u}):\epsilon(\mathbf{v}) d\Omega =
  \int_{\Omega} \textbf{f}^T\mathbf{v}d\Omega + \int_{\partial\Omega}\mathbf{t}^T\mathbf{v}dS,\\
  \text{onde} \quad \epsilon(\mathbf{v}) = sym(∇ \cdot \mathbf{v}).
$$

```python
from fenics import *
import numpy as np
import matplotlib.pyplot as plt

# ======= 1. Parâmetros gerais =======
L = 1.0
num_points_local = 200
num_points_eff = 400
r_min = 0.5
r_max = 2.0
p_i = 1.0
p_o = 0.0

# ======= 2. Malha e espaço de função para problemas locais (periódicos) =======
mesh_local = IntervalMesh(num_points_local, 0, L)

class PeriodicBoundary(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[0], 0.0) and on_boundary
    def map(self, x, y):
        y[0] = x[0] - L

V_local = FunctionSpace(mesh_local, 'P', 1, constrained_domain=PeriodicBoundary())

# ======= 3. Propriedades do material (2 fases) =======
class Lambda(UserExpression):
    def eval(self, value, x):
        value[0] = 10.0 if x[0] < 0.5*L else 20.0
    def value_shape(self):
        return ()

class Mu(UserExpression):
    def eval(self, value, x):
        value[0] = 5.0 if x[0] < 0.5*L else 15.0
    def value_shape(self):
        return ()

lambda_y = Lambda(degree=0)
mu_y     = Mu(degree=0)

# ======= 4. Problemas locais N11 e N12 =======
v = TestFunction(V_local)
u_trial = TrialFunction(V_local)

# N11
a_N11 = (lambda_y + 2*mu_y) * u_trial.dx(0) * v.dx(0) * dx
L_N11 = - (lambda_y + 2*mu_y) * Constant(1.0) * v.dx(0) * dx
N11 = Function(V_local)
solve(a_N11 == L_N11, N11)

# N12
a_N12 = (lambda_y + 2*mu_y) * u_trial.dx(0) * v.dx(0) * dx
L_N12 = - lambda_y * Constant(1.0) * v.dx(0) * dx
N12 = Function(V_local)
solve(a_N12 == L_N12, N12)

# ======= 5. Médias periódicas =======
C_eff  = assemble((lambda_y + 2*mu_y)*(1 + N11.dx(0)) * dx) / L
C_circ = assemble(((lambda_y + 2*mu_y)*N12.dx(0) + lambda_y) * dx) / L
print("C_eff =", float(C_eff), " C_circ =", float(C_circ))

# ======= 6. Verificações de periodicidade =======
y0 = 0.0
yL = L
eps = L/num_points_local
print("Periodicidade N11:", abs(N11(yL)-N11(y0)))
print("Periodicidade N12:", abs(N12(yL)-N12(y0)))

# ======= 7. Problema homogenizado =======
mesh_eff = IntervalMesh(num_points_eff, r_min, r_max)
V_eff = FunctionSpace(mesh_eff, 'P', 1)

u = TrialFunction(V_eff)
v = TestFunction(V_eff)
r = SpatialCoordinate(mesh_eff)[0]

# Forma fraca com peso r em todos os termos
a_eff = C_eff * u.dx(0) * v.dx(0) * r * dx + C_circ * u * v * r * dx
L_eff = Constant(0.0) * v * dx

# ======= 8. Fronteiras e pressões Neumann fracas =======
boundary_markers = MeshFunction('size_t', mesh_eff, mesh_eff.topology().dim()-1, 0)

class InnerBoundary(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[0], r_min) and on_boundary

class OuterBoundary(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[0], r_max) and on_boundary

inner = InnerBoundary()
outer = OuterBoundary()
inner.mark(boundary_markers, 1)
outer.mark(boundary_markers, 2)
ds = Measure('ds', domain=mesh_eff, subdomain_data=boundary_markers)

L_eff += -p_i * v * ds(1)
L_eff += -p_o * v * ds(2)

# Resolver
u_eff = Function(V_eff)
solve(a_eff == L_eff, u_eff)

# ======= 9. Verificação de tensões radiais =======
h = 1e-8
du_dr_min = (u_eff(r_min + h) - u_eff(r_min)) / h
du_dr_max = (u_eff(r_max) - u_eff(r_max - h)) / h
sigma_rr_min = C_eff * du_dr_min + C_circ * u_eff(r_min) / r_min
sigma_rr_max = C_eff * du_dr_max + C_circ * u_eff(r_max) / r_max
print("sigma_rr(r_min) =", float(sigma_rr_min), "  (esperado ≈ -p_i =", -p_i, ")")
print("sigma_rr(r_max) =", float(sigma_rr_max), "  (esperado ≈ -p_o =", -p_o, ")")

# ======= 10. Plots =======
plt.figure()
plot(N11)
plt.title("N11(y)")
plt.figure()
plot(N12)
plt.title("N12(y)")
plt.figure()
plot(u_eff)
plt.title("u_r^0(r) homogenizado")
plt.show()
```

```python
"""
Análise de homogeneização para materiais compostos usando FEniCS
Resolve problemas locais periódicos e problema homogenizado em coordenadas cilíndricas
"""
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple, Dict

from fenics import *


class MaterialProperties:
    """Classe para encapsular propriedades do material bifásico"""

    def __init__(self, L: float):
        self.L = L
        self.lambda_y = self.LambdaExpression(L)
        self.mu_y = self.MuExpression(L)

    class LambdaExpression(UserExpression):
        def __init__(self, L, **kwargs):
            super().__init__(**kwargs)
            self.L = L

        def eval(self, value, x):
            value[0] = 10.0 if x[0] < 0.5 * self.L else 20.0

        def value_shape(self):
            return ()

    class MuExpression(UserExpression):
        def __init__(self, L, **kwargs):
            super().__init__(**kwargs)
            self.L = L

        def eval(self, value, x):
            value[0] = 5.0 if x[0] < 0.5 * self.L else 15.0

        def value_shape(self):
            return ()


class PeriodicBoundary(SubDomain):
    """Condição de contorno periódica"""

    def __init__(self, L):
        super().__init__()
        self.L = L

    def inside(self, x, on_boundary):
        return near(x[0], 0.0) and on_boundary

    def map(self, x, y):
        y[0] = x[0] - self.L


class HomogenizationSolver:
    """Solver principal para análise de homogeneização"""

    def __init__(self, config: Dict):
        """
        Inicializa o solver com parâmetros de configuração

        Args:
            config: Dicionário com parâmetros do problema
        """
        self.config = config
        self.results = {}

        # Extrair parâmetros
        self.L = config['L']
        self.num_points_local = config['num_points_local']
        self.num_points_eff = config['num_points_eff']
        self.r_min = config['r_min']
        self.r_max = config['r_max']
        self.p_i = config['p_i']
        self.p_o = config['p_o']

        # Inicializar propriedades do material
        self.material = MaterialProperties(self.L)

        # Configurar malhas e espaços de função
        self._setup_meshes_and_spaces()

    def _setup_meshes_and_spaces(self):
        """Configura malhas e espaços de função"""
        # Malha local (periódica)
        self.mesh_local = IntervalMesh(self.num_points_local, 0, self.L)
        periodic_bc = PeriodicBoundary(self.L)
        self.V_local = FunctionSpace(
            self.mesh_local, 'P', 1,
            constrained_domain=periodic_bc
        )

        # Malha efetiva (homogenizada)
        self.mesh_eff = IntervalMesh(self.num_points_eff, self.r_min, self.r_max)
        self.V_eff = FunctionSpace(self.mesh_eff, 'P', 1)

    def solve_local_problems(self) -> Tuple[Function, Function]:
        """
        Resolve os problemas locais N11 e N12

        Returns:
            Tupla com as soluções (N11, N12)
        """
        print("Resolvendo problemas locais...")

        # Funções teste e trial
        v = TestFunction(self.V_local)
        u_trial = TrialFunction(self.V_local)

        # Material composto
        lambda_y = self.material.lambda_y
        mu_y = self.material.mu_y

        # Problema N11
        print("  Resolvendo N11...")
        a_N11 = (lambda_y + 2*mu_y) * u_trial.dx(0) * v.dx(0) * dx
        L_N11 = -(lambda_y + 2*mu_y) * v.dx(0) * dx

        N11 = Function(self.V_local)
        solve(a_N11 == L_N11, N11)

        # Problema N12
        print("Resolvendo N12...")
        a_N12 = (lambda_y + 2*mu_y) * u_trial.dx(0) * v.dx(0) * dx
        L_N12 = -lambda_y * v.dx(0) * dx

        N12 = Function(self.V_local)
        solve(a_N12 == L_N12, N12)

        eps = self.L / (10 * self.num_points_local) # Parâmetro geométrico pequeno
        y0 = eps
        yL = self.L - eps
        y_vals = np.linspace(0.01, yL, 100)
        N11_vals = [N11(y) for y in y_vals]
        N12_vals = [N12(y) for y in y_vals]

        self.results['N11'] = N11_vals
        self.results['N12'] = N12_vals

        return N11, N12

    def compute_effective_coefficients(self, N11: Function, N12: Function) -> Tuple[float, float]:
        """
        Calcula os coeficientes efetivos

        Args:
            N11, N12: Soluções dos problemas locais

        Returns:
            Tupla com (C_eff, C_circ)
        """
        print("Calculando coeficientes efetivos...")

        lambda_y = self.material.lambda_y
        mu_y = self.material.mu_y
        y = SpatialCoordinate(self.mesh_local)[0]

        # Coeficientes homogeneizados
        # C_eff = assemble((lambda_y + 2*mu_y) * (1 + N11.dx(0)) * y * dx) / self.L
        # C_circ = assemble((lambda_y + 2*mu_y) * (1 + N12.dx(0)) * y * dx) / self.L
        C_eff = assemble((lambda_y + 2*mu_y) * (1 + N11.dx(0)) * y * dx) / self.L
        C_circ = assemble((lambda_y + 2*mu_y) * (1 + N12.dx(0)) * y * dx) / self.L

        C_eff_val = float(C_eff)
        C_circ_val = float(C_circ)

        print(f"  C_eff = {C_eff_val:.6f}")
        print(f"  C_circ = {C_circ_val:.6f}")

        self.results['C_eff'] = C_eff_val
        self.results['C_circ'] = C_circ_val

        return C_eff_val, C_circ_val

    def verify_periodicity(self, N11: Function, N12: Function) -> Dict[str, float]:
        """
        Verifica a periodicidade das soluções

        Args:
            N11, N12: Soluções dos problemas locais

        Returns:
            Dicionário com erros de periodicidade
        """
        print("Verificando periodicidade...")

        # Pontos nas extremidades (com tolerância numérica)
        eps = self.L / (10 * self.num_points_local)
        y0 = eps
        yL = self.L - eps

        error_N11 = abs(N11(yL) - N11(y0))
        error_N12 = abs(N12(yL) - N12(y0))

        print(f"  Erro periodicidade N11: {error_N11:.2e}")
        print(f"  Erro periodicidade N12: {error_N12:.2e}")

        periodicity_errors = {
            'N11_error': error_N11,
            'N12_error': error_N12
        }

        self.results['periodicity_errors'] = periodicity_errors
        return periodicity_errors

    def solve_homogenized_problem(self, C_eff: float, C_circ: float) -> Function:
        """
        Resolve o problema homogenizado

        Args:
            C_eff, C_circ: Coeficientes efetivos

        Returns:
            Solução u_eff
        """
        print("Resolvendo problema homogenizado...")

        # Funções teste e trial
        u = TrialFunction(self.V_eff)
        v = TestFunction(self.V_eff)
        r = SpatialCoordinate(self.mesh_eff)[0]

        # Forma bilinear (com peso r para coordenadas cilíndricas)
        # a_eff = C_eff * u.dx(0) * v.dx(0) * r * dx + C_circ * u * v * r * dx # Está incorreta, pois na equação homogeneizada temos apenas C_eff
        a_eff = 2 * C_eff * ( u / r**2 ) * v * dx - C_eff * u.dx(0) * v.dx(0) * dx
        L_eff = Constant(0.0) * v * dx

        # Configurar condições de contorno Neumann
        boundary_markers = self._setup_boundary_markers()
        ds = Measure('ds', domain=self.mesh_eff, subdomain_data=boundary_markers)

        # Aplicar pressões nas fronteiras
        L_eff += -self.p_i * v * ds(1)  # Fronteira interna
        L_eff += -self.p_o * v * ds(2)  # Fronteira externa

        # Resolver sistema
        u_eff = Function(self.V_eff)
        solve(a_eff == L_eff, u_eff)

        # Extrair a solução homogenizada
        r_vals = np.linspace(self.r_min, self.r_max, self.num_points_eff)
        u_vals = [u_eff(ri) for ri in r_vals]

        self.results['u_eff'] = u_vals

        return u_eff

    def _setup_boundary_markers(self):
        """Configura marcadores de fronteira"""
        boundary_markers = MeshFunction(
            'size_t', self.mesh_eff,
            self.mesh_eff.topology().dim() - 1, 0
        )

        class InnerBoundary(SubDomain):
            def __init__(self, r_min):
                super().__init__()
                self.r_min = r_min

            def inside(self, x, on_boundary):
                return near(x[0], self.r_min) and on_boundary

        class OuterBoundary(SubDomain):
            def __init__(self, r_max):
                super().__init__()
                self.r_max = r_max

            def inside(self, x, on_boundary):
                return near(x[0], self.r_max) and on_boundary

        inner = InnerBoundary(self.r_min)
        outer = OuterBoundary(self.r_max)

        inner.mark(boundary_markers, 1)
        outer.mark(boundary_markers, 2)

        return boundary_markers

    def verify_stress_conditions(self, u_eff: Function, C_eff: float, C_circ: float) -> Dict[str, float]:
        """
        Verifica as condições de tensão nas fronteiras

        Args:
            u_eff: Solução homogenizada
            C_eff, C_circ: Coeficientes efetivos

        Returns:
            Dicionário com tensões calculadas
        """
        print("Verificando condições de tensão...")

        # Usar diferenças finitas mais precisas
        h = (self.r_max - self.r_min) / (100 * self.num_points_eff)

        # Derivadas numéricas
        du_dr_min = (u_eff(self.r_min + h) - u_eff(self.r_min)) / h
        du_dr_max = (u_eff(self.r_max) - u_eff(self.r_max - h)) / h

        # Tensões radiais
        sigma_rr_min = C_eff * du_dr_min + C_circ * u_eff(self.r_min) / self.r_min
        sigma_rr_max = C_eff * du_dr_max + C_circ * u_eff(self.r_max) / self.r_max

        sigma_rr_min_val = float(sigma_rr_min)
        sigma_rr_max_val = float(sigma_rr_max)

        print(f"  σ_rr(r_min) = {sigma_rr_min_val:.6f} (esperado ≈ {-self.p_i:.6f})")
        print(f"  σ_rr(r_max) = {sigma_rr_max_val:.6f} (esperado ≈ {-self.p_o:.6f})")

        # Calcular erros relativos
        error_min = abs(sigma_rr_min_val - (-self.p_i)) / abs(self.p_i) if self.p_i != 0 else abs(sigma_rr_min_val)
        error_max = abs(sigma_rr_max_val - (-self.p_o)) / abs(self.p_o) if self.p_o != 0 else abs(sigma_rr_max_val)

        print(f"  Erro relativo σ_rr(r_min): {error_min:.2e}")
        print(f"  Erro relativo σ_rr(r_max): {error_max:.2e}")

        stress_results = {
            'sigma_rr_min': sigma_rr_min_val,
            'sigma_rr_max': sigma_rr_max_val,
            'error_min': error_min,
            'error_max': error_max
        }

        self.results['stress_verification'] = stress_results
        return stress_results

    def verify_forces_equilibrium(): # Verificar se o equilbrio de forças foi satisfeito
      pass

    def run_complete_analysis(self):
        """Executa análise completa de homogeneização"""
        print("=" * 60)
        print("ANÁLISE DE HOMOGENEIZAÇÃO - MATERIAIS COMPOSTOS")
        print("=" * 60)

        # 1. Resolver problemas locais
        N11, N12 = self.solve_local_problems()

        # 2. Calcular coeficientes efetivos
        C_eff, C_circ = self.compute_effective_coefficients(N11, N12)

        # 3. Verificar periodicidade
        self.verify_periodicity(N11, N12)

        # 4. Resolver problema homogenizado
        u_eff = self.solve_homogenized_problem(C_eff, C_circ)

        # 5. Verificar condições de tensão
        self.verify_stress_conditions(u_eff, C_eff, C_circ)

        print("=" * 60)
        print("ANÁLISE CONCLUÍDA")
        print("=" * 60)

        return self.results

    def plot_results(self):
        """Gera gráficos dos resultados"""
        if not self.results:
            print("Execute a análise primeiro!")
            return

        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        fig.suptitle('Resultados da Análise de Homogeneização', fontsize=16)

        # Plot N11
        if 'N11' in self.results:
            axes[0,0].plot(self.results['N11'])
            axes[0,0].set_title('Função N11(y)')
            axes[0,0].set_xlabel('y')
            axes[0,0].set_ylabel('N11')
            axes[0,0].grid(True)

        # Plot N12
        if 'N12' in self.results:
            axes[0,1].plot(self.results['N12'], color = 'green')
            axes[0,1].set_title('Função N12(y)')
            axes[0,1].set_xlabel('y')
            axes[0,1].set_ylabel('N12')
            axes[0,1].grid(True)

        # # Plot solução homogenizada
        if 'u_eff' in self.results:
            axes[1,0].plot(self.results['u_eff'])
            axes[1,0].set_title('Deslocamento u_r^0(r)')
            axes[1,0].set_xlabel('r')
            axes[1,0].set_ylabel('u_r^0')
            axes[1,0].grid(True)

        # Plot tensões
        # print(self.results['C_eff'])
        # print(self.results['C_circ'])
        # print(self.results['u_eff'])
        # r_vals = np.linspace(self.r_min, self.r_max, self.num_points_eff)
        # stress_results = self.results['C_eff'] * self.results['u_eff'] + self.results['C_circ'] * self.results['u_eff'] / r_vals
        # axes[1,1].plot(r_vals, stress_results, color = 'red')

        # Plot propriedades do material
        y_vals = np.linspace(0, self.L, 100)
        lambda_vals = [10.0 if y < 0.5*self.L else 20.0 for y in y_vals]
        mu_vals = [5.0 if y < 0.5*self.L else 15.0 for y in y_vals]

        axes[1,1].plot(y_vals, lambda_vals, 'b-', label='λ(y)', linewidth=2)
        axes[1,1].plot(y_vals, mu_vals, 'r-', label='μ(y)', linewidth=2)
        axes[1,1].set_title('Propriedades do Material')
        axes[1,1].set_xlabel('y')
        axes[1,1].set_ylabel('Valor')
        axes[1,1].legend()
        axes[1,1].grid(True)

        plt.tight_layout()
        plt.show()

"""Função principal"""
# Configuração do problema
config = {
    'L': 1.0,                    # Comprimento da célula unitária
    'num_points_local': 200,     # Pontos na malha local
    'num_points_eff': 400,       # Pontos na malha efetiva
    'r_min': 0.5,               # Raio interno
    'r_max': 2.0,               # Raio externo
    'p_i': 1.0,                 # Pressão interna
    'p_o': 0.0                  # Pressão externa
}

# Criar e executar solver
solver = HomogenizationSolver(config)
results = solver.run_complete_analysis()

# Gerar gráficos
solver.plot_results()
```

# Proposta do Perplexity

```python
import numpy as np
import matplotlib.pyplot as plt
import logging
from dataclasses import dataclass
from typing import Optional
from fenics import *

# Configuração do logging
def setup_logging():
    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )

setup_logging()

@dataclass
class HomogenizationParams:
    L: float = 1.0
    num_points_local: int = 200
    num_points_eff: int = 400
    r_min: float = 0.5
    r_max: float = 2.0
    p_i: float = 1.0
    p_o: float = 0.0
    lambda_1: float = 10.0
    mu_1: float = 5.0
    lambda_2: float = 20.0
    mu_2: float = 15.0

class PeriodicBoundary(SubDomain):
    def __init__(self, L):
        super().__init__()
        self.L = L
    def inside(self, x, on_boundary):
        return near(x[0], 0.0) and on_boundary
    def map(self, x, y):
        y[0] = x[0] - self.L

def safe_divide_by_r(u, r, eps=1e-12):
    """Divisão segura por r para evitar singularidades em r=0"""
    return conditional(gt(r, eps), u / r, u.dx(0))

def solve_local_problems(params: HomogenizationParams):
    mesh_local = IntervalMesh(params.num_points_local, 0, params.L)
    V_local = FunctionSpace(mesh_local, 'P', 2, constrained_domain=PeriodicBoundary(params.L))

    materials = MeshFunction('size_t', mesh_local, mesh_local.topology().dim())
    materials.set_all(0)
    class Material2(SubDomain):
        def inside(self, x, on_boundary):
            return x[0] >= 0.5 * params.L
    Material2().mark(materials, 1)

    V_DG0 = FunctionSpace(mesh_local, 'DG', 0) # O que é DG0
    lambda_y_func = Function(V_DG0)
    mu_y_func = Function(V_DG0)

    for cell in cells(mesh_local):
        if materials[cell] == 0:
            lambda_y_func.vector()[cell.index()] = params.lambda_1
            mu_y_func.vector()[cell.index()] = params.mu_1
        else:
            lambda_y_func.vector()[cell.index()] = params.lambda_2
            mu_y_func.vector()[cell.index()] = params.mu_2

    v = TestFunction(V_local)
    u_trial = TrialFunction(V_local)
    a = (lambda_y_func + 2 * mu_y_func) * dot(grad(u_trial), grad(v)) * dx

    L_N11 = - dot((lambda_y_func + 2 * mu_y_func), v.dx(0)) * dx
    N11 = Function(V_local)
    solve(a == L_N11, N11, solver_parameters={'linear_solver': 'mumps'})

    L_N12 = - dot(lambda_y_func, v.dx(0)) * dx
    N12 = Function(V_local)
    solve(a == L_N12, N12, solver_parameters={'linear_solver': 'mumps'})

    return N11, N12, materials

def compute_effective_properties(N11, N12, materials, params: HomogenizationParams):
    mesh_local = N11.function_space().mesh()
    V_DG0 = FunctionSpace(mesh_local, 'DG', 0)
    lambda_y = Function(V_DG0)
    mu_y = Function(V_DG0)
    for cell in cells(mesh_local):
        mat_id = materials[cell]
        lambda_y.vector()[cell.index()] = [params.lambda_1, params.lambda_2][mat_id]
        mu_y.vector()[cell.index()] = [params.mu_1, params.mu_2][mat_id]
    C_eff = assemble((lambda_y + 2 * mu_y) * (1 + N11.dx(0)) * dx) / params.L
    # C_circ = assemble(((lambda_y + 2 * mu_y) * N12.dx(0) + lambda_y) * dx) / params.L
    C_circ = assemble((lambda_y + 2 * mu_y) * (1 + N12.dx(0)) * dx) / params.L # Problema local corrigido
    return C_eff, C_circ, None

def solve_homogenized_problem(C_eff, C_circ, params: HomogenizationParams):
    mesh_eff = IntervalMesh(params.num_points_eff, params.r_min, params.r_max)
    V_eff = FunctionSpace(mesh_eff, 'P', 2)
    u = TrialFunction(V_eff)
    v = TestFunction(V_eff)
    r = SpatialCoordinate(mesh_eff)[0]

    a = C_eff * dot(grad(u), grad(v)) * r * dx + C_circ * u * v * r * dx # Verificaar essa forma variacional do problema homogeneizado
    L = Constant(0.0) * v * r * dx

    boundary_markers = MeshFunction('size_t', mesh_eff, mesh_eff.topology().dim() - 1, 0)
    class InnerBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return near(x[0], params.r_min) and on_boundary
    class OuterBoundary(SubDomain):
        def inside(self, x, on_boundary):
            return near(x[0], params.r_max) and on_boundary

    inner = InnerBoundary()
    outer = OuterBoundary()
    inner.mark(boundary_markers, 1)
    outer.mark(boundary_markers, 2)
    ds = Measure('ds', domain=mesh_eff, subdomain_data=boundary_markers)

    L -= params.p_i * v * ds(1)
    L -= params.p_o * v * ds(2)

    u_eff = Function(V_eff)
    solve(a == L, u_eff, solver_parameters={'linear_solver': 'mumps'})
    return u_eff

def validate_solution(u_eff, params: HomogenizationParams, C_eff, C_circ):
    mesh_eff = u_eff.function_space().mesh()
    r = SpatialCoordinate(mesh_eff)[0]

    du_dr = u_eff.dx(0)
    u_over_r = u_eff / r
    sigma_rr_expr = C_eff * du_dr + C_circ * u_over_r

    V_eval = FunctionSpace(mesh_eff, 'P', 1)
    sigma_rr = project(sigma_rr_expr, V_eval)
    sigma_rr_min = sigma_rr(params.r_min)
    sigma_rr_max = sigma_rr(params.r_max)

    validation = {
        'sigma_rr_r_min': (sigma_rr_min),
        'sigma_rr_r_max': (sigma_rr_max),
        'expected_p_min': -params.p_i,
        'expected_p_max': -params.p_o
    }
    return validation

def plot_results(N11, N12, u_eff, params: HomogenizationParams):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    eps = params.L / (10 * params.num_points_local) # Parâmetro geométrico pequeno
    y0 = eps
    yL = params.L - eps
    y_vals = np.linspace(0.01, yL, 100)
    N11_vals = [N11(y) for y in y_vals]
    N12_vals = [N12(y) for y in y_vals]
    axes[0].plot(y_vals, N11_vals, 'b-', label='N11(y)', linewidth=2)
    axes[1].plot(y_vals, N12_vals, 'r-', label='N12(y)', linewidth=2)
    # plt.plot(N11)
    # axes[0].set_title("N11(y)")
    # axes[0].set_xlabel('y')
    # axes[0].set_ylabel('N11')
    # axes[0].grid(True)

    # plt.plot(N12)
    # axes[1].set_title("N12(y)")
    # axes[1].set_xlabel('y')
    # axes[1].set_ylabel('N12')
    # axes[1].grid(True)

    r_vals = np.linspace(params.r_min, params.r_max, 200)
    u_vals = [u_eff(r) for r in r_vals]
    axes[2].plot(r_vals, u_vals, 'b-', linewidth=2)
    axes[2].set_title("Deslocamento radial homogenizado")
    axes[2].set_xlabel('r')
    axes[2].set_ylabel('u_r^0(r)')
    axes[2].grid(True)

    plt.tight_layout()
    plt.show()

params = HomogenizationParams()
try:
    logging.info("Iniciando resolução dos problemas locais...")
    N11, N12, materials = solve_local_problems(params)
    logging.info("Calculando propriedades efetivas...")
    C_eff, C_circ, _ = compute_effective_properties(N11, N12, materials, params)
    logging.info(f"Resolvendo problema homogenizado com C_eff={C_eff:.6f}, C_circ={C_circ:.6f}")
    u_eff = solve_homogenized_problem(C_eff, C_circ, params)
    logging.info("Validando solução...")
    validation = validate_solution(u_eff, params, C_eff, C_circ)

    print("Tensão radial em r_min:", validation['sigma_rr_r_min'], "( esperado:", validation['expected_p_min'], ")")
    print("Tensão radial em r_max:", validation['sigma_rr_r_max'], "( esperado:", validation['expected_p_max'], ")")

    plot_results(N11, N12, u_eff, params)

    results = {
      'N11': N11,
      'N12': N12,
      'u_eff': u_eff,
      'C_eff': C_eff,
      'C_circ': C_circ,
      'validation': validation
    }

except Exception as e:
    logging.error(f"Erro na execução: {e}")
    raise
```

```python
from fenics import *

# Create a simple 1D mesh
mesh = IntervalMesh(10, 0, 1)

# Define subdomains using MeshFunction
materials = MeshFunction("size_t", mesh, mesh.topology().dim())
materials.set_all(0) # Default subdomain

class Subdomain1(SubDomain):
    def inside(self, x, on_boundary):
        return x[0] < 0.5

class Subdomain2(SubDomain):
    def inside(self, x, on_boundary):
        return x[0] >= 0.5

Subdomain1().mark(materials, 1)
Subdomain2().mark(materials, 2)

# Define a spatially varying coefficient using a DG0 Function
V_DG0 = FunctionSpace(mesh, 'DG', 0)
spatial_coeff = Function(V_DG0)

# Assign values to the DG0 function based on subdomain markers
coeff_value_1 = 5.0
coeff_value_2 = 10.0
for cell in cells(mesh):
    materials_id = materials[cell.index()] # Get the subdomain ID for the cell
    if materials_id == 1:
        spatial_coeff.vector()[cell.index()] = coeff_value_1
    elif materials_id == 2: # Use the correct subdomain ID
        spatial_coeff.vector()[cell.index()] = coeff_value_2
    else: # Handle the default case if necessary
        spatial_coeff.vector()[cell.index()] = coeff_value_1 # Assign default value or raise an error

# Define the integration measure with subdomain data (optional in this case, but good practice)
# dx_subdomains = Measure('dx', domain=mesh, subdomain_data=materials)
# We can just use the standard dx measure when using a DG0 function

# Attempt to assemble a simple integral with the spatially varying coefficient
try:
    # Use the DG0 function directly in the assemble call
    integral_value = assemble(spatial_coeff * dx)
    print("Successfully assembled the integral:", integral_value)
except Exception as e:
    print("Error during assembly of the simplified integral:", e)
```

# Método por domínio bidimensional circular

```python
from fenics import *
from mshr import *
import numpy as np
import matplotlib.pyplot as plt

Re = 11.
Ri = 9.
num_layers = 10  # Número de camadas
# Propriedades dos materiais
E1 = 210e9  # Módulo de Young do material 1 (Pa)
nu1 = 0.3   # Coeficiente de Poisson do material 1
E2 = 70e9   # Módulo de Young do material 2 (Pa)
nu2 = 0.33  # Coeficiente de Poisson do material 2

rect = Rectangle(Point(0., 0.), Point(Re, Re))
domain = Circle(Point(0., 0.), Re, 100) - Circle(Point(0., 0.), Ri, 100)
domain = domain - Rectangle(Point(0., -Re), Point(-2*Re, Re)) \
                - Rectangle(Point(0., 0.), Point(Re, -Re))

mesh = generate_mesh(domain, 100)
plot(mesh)

class Bottom(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[1], 0) and on_boundary
class Left(SubDomain):
    def inside(self, x, on_boundary):
        return near(x[0], 0) and on_boundary
class Outer(SubDomain):
    def inside(self, x, on_boundary):
        return near(sqrt(x[0]**2+x[1]**2), Re, 1e-1) and on_boundary

facets = MeshFunction("size_t", mesh, 1)
facets.set_all(0)
Bottom().mark(facets, 1)
Left().mark(facets, 2)
Outer().mark(facets, 3)
ds = Measure("ds", subdomain_data=facets)

class MaterialProperties():
    def __init__(self, E1, nu1, E2, nu2, R_inner, R_outer, num_layers, **kwargs):
        #super().__init__(**kwargs)
        self.E1 = E1
        self.nu1 = nu1
        self.E2 = E2
        self.nu2 = nu2
        self.R_inner = R_inner
        self.R_outer = R_outer
        self.num_layers = num_layers

    def eval(self, values, x):
        r = np.sqrt(x[0]**2 + x[1]**2)  # Coordenada radial
        layer_thickness = (self.R_outer - self.R_inner) / self.num_layers
        layer = int((r - self.R_inner) / layer_thickness)
        if layer % 2 == 0:
            values[0] = self.E1
            values[1] = self.nu1
        else:
            values[0] = self.E2
            values[1] = self.nu2

    def value_shape(self):
        return (2,)
material = MaterialProperties(E1, nu1, E2, nu2, Ri, Re, num_layers, degree=0)

x = SpatialCoordinate(mesh)

def epsilon(u):
    return 0.5*(nabla_grad(u) + nabla_grad(u).T)
    #return sym(nabla_grad(u))

# E = Constant(2e5) #MPa
# nu = Constant(0.3)
# mu = E/2/(1+nu)
# lmbda = E*nu/(1+nu)/(1-2*nu)

def sigma(u):
    E = material.E1
    nu = material.E2
    lmbda = E * nu / ((1 + nu) * (1 - 2 * nu))  # Coeficiente de Lamé para deformação plana
    mu = E / (2 * (1 + nu))  # Coeficiente de Lamé (módulo de cisalhamento)
    return lmbda * tr(epsilon(u)) * Identity(2) + 2 * mu * epsilon(u)

n = FacetNormal(mesh)
p = Constant(10.)

V = VectorFunctionSpace(mesh, 'CG', degree=2)
du = TrialFunction(V)
u_ = TestFunction(V)
a = inner(sigma(du), epsilon(u_))*x[0]*dx
l = inner(-p*n, u_)*x[0]*ds(3) #Por que a integração foi feita na superfície externa? Por que a pressão é aplicada nessa superfície?

u = Function(V, name="Displacement")

bcs = [DirichletBC(V.sub(1), Constant(0), facets, 1),
       DirichletBC(V.sub(0), Constant(0), facets, 2)]

solve(a == l, u, bcs)
```

```python
print("Inwards radial displacement at (r=Re, theta=0): \
{:1.7f} (FE) {:1.7f} (Exact)".format(-u(Re, 0.)[0], float(Re**3/(Re**3-Ri**3)*((1-2*nu)*Re+(1+nu)*Ri**3/2/Re**2)*p/E)))
print("Inwards radial displacement at (r=Ri, theta=0): \
{:1.7f} (FE) {:1.7f} (Exact)".format(-u(Ri, 0.)[0], float(Re**3/(Re**3-Ri**3)*((1-2*nu)*Ri+(1+nu)*Ri/2)*p/E)))
```
